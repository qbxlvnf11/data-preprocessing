```python
import PIL.Image as pilimg
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
path = './test.jpeg'
```

## - PIL Read Image

```python
# Read image
pil_img = pilimg.open(path)
```

## - Read Numpy (cv2) Image

```python
# Read image
cv_img = cv2.imread(path, cv2.IMREAD_COLOR)
```

```python
# PIL to Numpy
np_img = np.array(pil_img)
```

## - Image Size: PIL, Numpy

```python
# PIL: W, H
width, height = pil_img.size
print('Width:', width)
print('Height;', height)
```

```python
# np (cv2): H, W, C
height, width, channel = np_img.shape
print('Width:', width)
print('Height:', height)
print('Channel:', channel)
```

## - Show Image

```python
# PIL display image
display(pil_img)
```

```python
# Numpy display image
plt.imshow(np_img, interpolation='nearest')
plt.show()
```
