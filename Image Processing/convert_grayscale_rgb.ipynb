```python
from matplotlib import pyplot as plt
from PIL import Image
import cv2
import numpy as np
```

```python
path = './sample_img/nm0000188_rm2570898432_1945-8-14_1979.jpg'
```

### - PIL

```python
# Read grayscale img
gray_image = Image.open(path)
```

```python
# Shape
np.array(gray_image).shape
```

```python
# Show img
gray_image
```

```python
# Convert from gray to rgb
rgb_image = gray_image.convert("RGB")
```

```python
# Shape
np.array(rgb_image).shape
```

```python
# Show img
rgb_image
```

## - OpenCV

```python
# Read grayscale img
gray_image = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
```

```python
# Shape
gray_image.shape
```

```python
gray_image
```

```python
# Show img
plt.imshow(gray_image)
plt.title('gray_image')
plt.show()
```

```python
# Convert from gray to rgb
rgb_image = cv2.cvtColor(gray_image, cv2.COLOR_GRAY2RGB)
```

```python
# Shape
rgb_image.shape
```

```python
rgb_image
```

```python
# Show img
plt.imshow(rgb_image)
plt.title('rgb_image')
plt.show()
```
