- Load an audio file as a floating point time series
    - Sample rate (sr): The (positive integer) number of samples per second of a time series (defalut = 22050).

```python
y, sr = librosa.load('test.wav', sr=16000)
```

- STFT (Short-time Fourier Transfrom)
    - Hop length (hop_length): number audio of frames between STFT columns, The number of samples between successive frames, e.g., the columns of a spectrogram.
    - Window length (n_fft): FFT window size. The length (width) of the window function (e.g., Hann window). Note that this can be smaller than the frame length used in a short-time Fourier transform.

```python
# Short-time Fourier transform (STFT)
# hop_length = number audio of frames between STFT columns, n_fft = FFT window size
hop_length = 1000
n_fft = 2500
Y = librosa.stft(y, n_fft=n_fft, hop_length=hop_length)
```

- Visualization

```python
from matplotlib import pyplot as plt
import librosa.display
import numpy as np
```

```python
cmap_lst = [plt.cm.rainbow, plt.cm.Blues, plt.cm.autumn, plt.cm.RdYlGn]
```

```python
D = np.abs(Y)
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), y_axis='linear', x_axis='time')
plt.title('Dual Tone')
# plt.ylim(0, 4000)
plt.show()
```
